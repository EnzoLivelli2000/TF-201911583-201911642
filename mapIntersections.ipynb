```python
import folium
import pandas as pd
import geopandas as gpd
import numpy as np
```

```python
limaIntersecciones = gpd.read_file('Lima-intersecciones_500.csv')
coordenadasOrigen = limaIntersecciones[["Latitud_Origen_Interseccion","Longitud_Origen_Interseccion"]]
cantidadCoordenadas = coordenadasOrigen["Latitud_Origen_Interseccion"].count()
listCoordenadasOrigen = []

for x in range(cantidadCoordenadas):
    tupla = (float(coordenadasOrigen["Latitud_Origen_Interseccion"][x]),float(coordenadasOrigen["Longitud_Origen_Interseccion"][x]))
    listCoordenadasOrigen.append(tupla)
    
print(listCoordenadasOrigen[0])
```

```python
coordenadasDestino = limaIntersecciones[["Latitud_Destino_Interseccion","Longitud_Destino_Interseccion"]]
listCoordenadasDestino = []

for x in range(cantidadCoordenadas):
    tupla = (float(coordenadasDestino["Latitud_Destino_Interseccion"][x]),float(coordenadasDestino["Longitud_Destino_Interseccion"][x]))
    listCoordenadasDestino.append(tupla)
    
print(listCoordenadasDestino[0])
```

```python
# m = folium.Map(location=[-12.211152, -77.015638], zoom_start=14)
m = folium.Map()
listaCoordenadas=[]
for x in range(cantidadCoordenadas):
    listaCoordenadas.append([listCoordenadasOrigen[x],listCoordenadasDestino[x]])
m.fit_bounds(listaCoordenadas)
```

```python
listaCoordenadas
```

```python
m.save("mapacasa.html")
```

```python
tooltip = "Haz click"
for x in range(cantidadCoordenadas):
    var = listaCoordenadas[x][0]
    folium.Marker(listaCoordenadas[x][0], popup=var, tooltip=tooltip, icon=folium.Icon(color='red',icon='info-sign')).add_to(m)
m.save("mapacasa.html")
```

```python
folium.PolyLine(
    locations = listaCoordenadas,
    popup = "Zona de paseo"
).add_to(m)
m.save("mapacasa.html")
```
